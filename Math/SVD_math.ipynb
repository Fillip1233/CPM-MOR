## 1.SVD-MOR
### 1. Construction of H (s)
$G=QJQ^T$ is the $LDL^T$ decomposition of the **symmetic** $G$ matrix(our matrix is not symmetic, so we use LU decomposition), $J$ is a simple matrix 
$$H(s) = N^T(G+sC)^{-1}M \\
= N^T[Q(J+sQ^{-1}CQ^{-T})Q^T]^{-1}M\\
= N^TQ^{-T}(J+sA)^{-1}Q^{-1}M$$
$A = Q^{-1}CQ^{-T}$, $L = Q^{-1}N$ and $R=Q^{-1}M$
construct $B=[LR]$, where $B \in \mathbf{R}^{n \times (m=p+q)}$, $p$ and $q$ are the number of columns of the input and output matrix

### 2. Use LU decomposition
$G = LJU$ then:
$$H(s) = N^T(G+sC)^{-1}M \\
= N^T[L(J+sL^{-1}CU^{-1})U]^{-1}M\\
= N^TU^{-1}(J+sA)^{-1}L^{-1}M$$
$A = L^{-1}CU^{-1}$, $L = U^{-T}N$ and $R=L^{-1}M$ construct $B=[LR]$, where $B \in \mathbf{R}^{n \times (m=p+q)}$, $p$ and $q$ are the number of columns of the input and output matrix

$L = (BE_L)$ and $R = (BE_R)$

### 3. MOR part
$$H(s) = E_L^TB^T(J+sA)^{-1}BE_R$$
then, perform svd decomposition on B: $ B = U \Sigma V^T \approx U_r \Sigma_r V_r^T$ \
$H(s)$ then become $H(s) = E_L^T\Sigma_r V_r U_r^T(J+sA)^{-1}U_r \Sigma_r V_r^TE_R$ \
we use $H_r(s) = U_r^T(J+sA)^{-1}U_r$ for mor
$$ H_r(s) = U_r^T(J+sA)^{-1}U_r \\
   = U_r^T(J+sL^{-1}CU^{-1})^{-1}U_r \\
   = U_r^T[L^{-1}(G+sC)U^{-1}]^{-1}U_r \\
   = U_r^TU(G+sC)LU_r
$$
Finally $$H(s) = E_L^T \Sigma_r V_r U_r^TU(G+sC)^{-1}LU_r \Sigma_r V_r^TE_R$$

## 4. Another solution
$H(s) = N^T(G+sC)^{-1}M = N^T(I+sG^{-1}C)^{-1}G^{-1}M$ \
Then $B=[N G^{-1}M]\approx U_r \Sigma_r V_r^T$
$$H(s) = E_L^TB^T(I+sG^{-1}C)^{-1}BE_R \approx E_L^TV_r\Sigma_r U_r^T(I+sG^{-1}C)^{-1}U_r\Sigma_r V_r^TE_R$$
Finially, perform MOR on $U_r^T(I+sG^{-1}C)^{-1}U_r$
